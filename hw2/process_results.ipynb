```python
import plot as p
import matplotlib.pyplot as plt
```

```python
## SA: EVALUATE FAST ANNEALING

file_path = 'test_sa_T10_alphaFASTANN_steps1000.txt'
t1 = p.process_test_perf(file_path)
file_path = 'test_sa_T100_alphaFASTANN_steps1000.txt'
t2 = p.process_test_perf(file_path)
file_path = 'test_sa_T1000_alphaFASTANN_steps1000.txt'
t3 = p.process_test_perf(file_path)


fig, axes = plt.subplots(1,3, figsize = (15,5), sharey='all')

for i, perf_list in enumerate(t1):
    axes[0].plot(perf_list, label=f'Test {i + 1}')

    axes[0].set_title("T10, Fast Annealing")
    axes[0].yaxis.set_major_locator(plt.MaxNLocator(8))  # Set the number of y-axis ticks

for i, perf_list in enumerate(t2):
    axes[1].plot(perf_list, label=f'Test {i + 1}')

    axes[1].set_title("T100, Fast Annealing")
    axes[1].yaxis.set_major_locator(plt.MaxNLocator(8))  # Set the number of y-axis ticks

for i, perf_list in enumerate(t3):
    axes[2].plot(perf_list, label=f'Test {i + 1}')

    axes[2].set_title("T1000, Fast Annealing")
    axes[2].yaxis.set_major_locator(plt.MaxNLocator(8))  # Set the number of y-axis ticks
    

for ax in axes:
    ax.set_xlabel("Step")
    ax.set_ylabel("Shortest Distance")
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
